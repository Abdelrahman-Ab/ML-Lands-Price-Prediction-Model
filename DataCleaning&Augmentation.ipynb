```python
import os
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.utils import img_to_array, load_img

# Directories for input and output
input_dir = 'C:\\Users\\peedo.abourayya\\Desktop\\Machine Learning\\Lands'
output_dir = 'C:\\Users\\peedo.abourayya\\Desktop\\Machine Learning\\preview2'
os.makedirs(output_dir, exist_ok=True)

# Define the data augmentation generator
datagen = ImageDataGenerator(
    rotation_range=40,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Function to apply augmentation to a single image
def augment_image(image_path, output_dir, augmentations_per_image=20):
    # Load the image
    img = load_img(image_path)  # Load the image
    x = img_to_array(img)       # Convert the image to a NumPy array
    x = x.reshape((1,) + x.shape)

    # Generate and save exactly 20 augmented images
    file_name = os.path.basename(image_path).split('.')[0]
    i = 0
    for batch in datagen.flow(x, batch_size=1,
                              save_to_dir=output_dir, save_prefix=file_name, save_format='jpeg'):
        i += 1
        if i >= augmentations_per_image:  # Stop after generating the specified number of augmentations
            break

# Process all images in the input directory
for file_name in os.listdir(input_dir):
    if file_name.lower().endswith(('.png', '.jpg', '.jpeg')):  # Filter for image files
        image_path = os.path.join(input_dir, file_name)
        augment_image(image_path, output_dir)

print(f"Augmented images saved to the directory: {output_dir}")
```
